## Generate Embeddings

```python
output = embeddings.embed_documents(texts)
```

```python
output[1]
```

## Push data to database

```python
index_name = "coffeeshop"

pc.create_index(
    name=index_name,
    dimension=768, # Replace with your model dimensions
    metric="cosine", # Replace with your model metric
    spec=ServerlessSpec(
        cloud="aws",
        region="us-east-1"
    ) 
)
```

```python
# Wait for the index to be ready
while not pc.describe_index(index_name).status['ready']:
    time.sleep(1)

index = pc.Index(index_name)

vectors = []
for text, e in zip(texts, output):
    entry_id = text.split(":")[0].strip()
    vectors.append({
        "id": entry_id,
        "values": e,
        "metadata": {'text': text}
    })
    
index.upsert(
    vectors=vectors,
    namespace="ns1"
)
```

## Get Closest documents

```python
query = "Is Cappuccino lactose-free?"

query_vector = embeddings.embed_query(query)

result = index.query(
    vector=query_vector,
    top_k=3,
    include_metadata=True,
    include_values=False,
    namespace="ns1"
)

for match in result["matches"]:
    print(f"Score: {match['score']:.2f}")
    print("Text:", match['metadata']['text'])
    print("---")
```

```python
from groq import Groq
import os
from dotenv import load_dotenv
load_dotenv()
api_key=os.getenv("GROQ_API_KEY")
def get_chatbot_response(messages, model_name="llama-3.1-8b-instant",  temperature=0.7):
    """
    Get response from a GROQ-hosted LLM (e.g., Mixtral, LLaMA3).

    Args:
        api_key: Your Groq API key.
        model_name: Model like 'mixtral-8x7b-32768', 'llama3-70b-8192', etc.
        messages: List of dicts -> [{"role": "user", "content": "Hello"}].
        temperature: Randomness level of output (0-1).

    Returns:
        Assistant's response text.
    """
    client = Groq()

    response = client.chat.completions.create(
        model=model_name,
        messages=messages,
        temperature=temperature,
        top_p=0.8,
        max_tokens=2048,
    )

    return response.choices[0].message.content.strip()
```

```python


messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "what is cpital of india?"}
]

reply = get_chatbot_response(messages)
print(reply)
```
